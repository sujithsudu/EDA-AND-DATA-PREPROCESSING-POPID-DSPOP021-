```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('police.csv')
df.head(5)
```

```python
df["violation_raw"].value_counts().head(5).plot(kind="bar")
```

```python
arrested = df[df["stop_outcome"] == "Arrest Driver"]
arrested["violation"].value_counts().head(5).plot(kind="bar")
```

```python
plt.figure(figsize=(8, 6))
plt.style.use("ggplot")
plt.title("Gender of people, stopped by the police.")
df["driver_gender"].value_counts().plot(kind="pie")
plt.show()
```
