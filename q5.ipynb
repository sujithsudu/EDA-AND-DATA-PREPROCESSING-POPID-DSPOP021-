```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv')
data.head()
```

```python
# Data Cleaning
ColumnToClean = ['total_cases', 'new_cases', 'total_deaths']
# filling missing values 
data[['continent']] = data[['continent']].fillna('')
data[ColumnToClean] = data[ColumnToClean].fillna(0)
data.loc[data['continent'] == "United Kingdom", "continent"] = "UK"
data.head()
```
